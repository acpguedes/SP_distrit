```python
import geopandas as gpd 
import pandas as pd
from shapely.geometry import Point
import matplotlib.pyplot as plt
import urllib
```

```python
import spdistrit.spdistrit as spd
```

```python
tt = spd.SpPolygon()
```

```python
tt.get_data()
```

```python
data = tt.load_distrites()
```

```python
data.plot()
```

```python
d = {'geometry':[Point(-46.66354, -23.53691)]}
point = gpd.GeoDataFrame(d)
```

```python
gpd.sjoin(data, point, how='right').Nome
```

```python
minx, miny, maxx, maxy = point.geometry.total_bounds
f, ax = plt.subplots()
point.plot(ax=ax, marker='o', color='red', markersize=20)
data.plot(ax=ax, color='green')
ax.set_xlim(minx - 0.2, maxx + 0.2)
ax.set_ylim(miny - 0.2, maxy + 0.2)
plt.show()
```
